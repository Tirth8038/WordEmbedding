# Word Embedding
Word embeddings provide a dense representation of words and their relative meanings.They are an improvement over sparse representations used in simpler bag of word model representations.Word embeddings can be learned from text data and reused among projects. They can also be learned as part of fitting a neural network on text data.



```python
#Importing Necessary Libraries for Tensorflow Hub
import numpy as np
import tensorflow_hub as hub
```

```python
#Loading the Dataset
embedding= hub.load ("https://tfhub.dev/google/Wiki-words-250/2")
```

```python
a = embedding ([" king "])
b = embedding ([" man "])
c = embedding ([" woman "])
d = embedding ([" queen "])
```

```python
# Storing the result ofvector(King) − vector(Man) + vector(Woman)
X= a-b+c
print("King-Man+Woman:", [X])
```

Finding Other Unique Similarities between Vector Representation of words like King, Man, Woman and Queen.

```python
# Storing the result ofvector(Woman) − vector(King) + vector(Queen) to find similarity of Man
Y= c-a+d
print("Woman-King+Queen:", [Y])
```

```python
# Storing the result of queen - woman + man to find similairy with vector King.
T=d-c+a
print("Queen-Woman+King:", [T])
```

```python
#Queen-Woman+Man to find similarity with King
P=d-c+b
print("Queen-Woman+Man:", [P])
```

```python
# Storing the result ofvector(King) − vector(Man) to find similarity with vector(Queen) − vector(Woman)
Z= a-b #King-Man
U= d-c #Queen-Woman
```

```python
#Calculating the Cosine Similarity between Vectors of Words.
from sklearn.metrics.pairwise import cosine_similarity

print("Similarity between King-Man+Woman and Queen:",cosine_similarity(X,d))
print("Similarity between Woman-King+Queen and Man:",cosine_similarity(Y,b))
print("Similarity between King-Man and Queen-Woman:",cosine_similarity(Z,U))
print("Similarity between Queen-Woman+King and Man:",cosine_similarity(T,b))
print("Similarity between Queen-Woman+Man and King:",cosine_similarity(P,a))

print("Similarity between Man and Woman:",cosine_similarity(b,c))
print("Similarity between King and Queen:",cosine_similarity(a,d))
print("Similarity between Man and King:",cosine_similarity(a,b))
print("Similarity between Woman and Queen:",cosine_similarity(c,d))
print("Similarity between Man and Queen:",cosine_similarity(b,d))
```

 To Find the top 5 closest words to the word "dog"

```python
# Creating my own small corpus of words related to dog
dog_corpus = ['pup', 'canine', 'doggo', 'pet', 'hound', 'puppy', 'sheep dog', 'bull dog', 'pet dog', 'cat','wolf']
```

```python
# Embedding all the words in the list from dog corpus
embed_dog = []
for i in range(len(dog_corpus)):
  embed_dog.append(embedding([dog_corpus[i]]))
```

```python
e=embedding(['dog'])
```

```python
#For Storing the Cosine Similarity of words
cos_sim = {}
for i in range(len(embed_dog)):
  cos_sim[dog_corpus[i]] = cosine_similarity(e,embed_dog[i])
```

```python
# For Creating Asceding order of Cosine Similarity
cos_sim_asc = {}
sorted_values = sorted(cos_sim.values())
for i in sorted_values:
    for k in cos_sim.keys():
        if cos_sim[k] == i:
            cos_sim_asc[k] = cos_sim[k]
            break
```

```python
# Top 5 words similar to dog
size = len(list(cos_sim_asc))
print(size)
top5 = list(cos_sim_asc.keys())
for i in range(size):
  if i<(size-5):
    continue
  else:
    print("Word :", top5[i])
    print("Cosine Similarity :", cos_sim_asc[top5[i]])
```
